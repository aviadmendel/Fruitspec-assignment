## part 1 - Fruit growth
### 1.1	What are the main factors that affect fruit size in citrus and what can be done to control the fruit size?

The main factors that can affect fruit size in citrus are:

1. <ins>Genetics (variety) -</ins> With the big developments we have today in genetics and breeding techniques, farmers have more options for growing species with bigger fruit size and higher yield. This aspect is less important when trying to answer the question of how to control the fruit size throughout the year.
2. <ins>Temperature -</ins> Another factor that affects a lot on fruit size is temperature. Generally speaking, the ideal temperature for citrus growing will be between 15 and 30 degrees, and to be even more precise when talking about Orri mandarin a temperature between 20-25 degrees will be ideal. Below 15 or above 30 degrees can lead to a negative effect on fruit size and also on quality and total yield. Another important note while talking about temperature is the importance of avoiding extreme temperature events and fluctuations in temperature Having a consistent temperature throughout the growing season will affect positively on the fruit's size and quality. This factor obviously can't be controlled by the farmer in the orchard like in some greenhouses.
3. <ins>Light -</ins> Light is a very important factor in the photosynthesis process of plants. As we know, the plant converts sunlight, water and carbon dioxide and turns them into glucose, which will affect the productivity of the tree and eventually the fruit size. When talking specifically about citrus, we can say that a good full sun of around 8 hours a day will be suitable. Same again as temperature - obviously we can't control the sunlight and clouds, (yet - with the developments today, who knows if in the future we will be able to control the weather) but we can affect the amount of sunlight that reaches the fruits. The way to do that is by pruning the trees which will lead to more sunlight reaching the fruits in the orchard (more about pruning in a later section). 
4. <ins>Irrigation -</ins> Irrigation is one of the most important factors in citrus trees growth and has a major impact on fruit size and It is important to keep the soil moist during the growing season. A young tree needs to be watered every couple of days, while a grown one that retains more water can be watered every couple of weeks (Of course, rain, temperature and soil type will affect the amount of water that should be given by irrigation). To little irrigation will lead to stress and a deficit, while too much irrigation can harm the roots and cause a deficit in oxygen, which is important to the growth of the tree. 
5. <ins>Fertilization -</ins>Another important tool the farmer has in hand is fertilization. NPK (nitrogen, phosphorus and potassium) are the three major elements in fertilizing and next to them are zinc, iron, copper and more. Fertilizing is an art that needs to consider the tree size, soil's fertility and availability of nutrients and the right amount of each nutrient for a citrus tree in our case. Again, either too little or too much fertilization can be detrimental to the trees, resulting in lower yields and smaller fruits.
6. <ins>Pruning and thinning -</ins> Pruning and thinning is another tactic to increase fruit size. Pruning involves the selective remove of branches and shoots from the tree and controls its size and shape and is usually done after harvest, while thinning involves the selective removal of fruits. One reason is that more energy from the tree goes to fewer fruits, leading to better quality and size (for example, more nutrients and water to each fruit). The second reason, which I already mentioned in another section, is to increase the availability of sunlight to the fruits.
<br>
<br>
To sum up, all the factors I mentioned can highly affect fruit size, but when talking about the control of fruit size throughout a growing season and the tools a farmer has in hand, the main three factors will be irrigation, fertilization, prunning and thinning.

## 1.2 Create a fruit growth curve for each tree from the data set 

```python
import pandas as pd
import matplotlib.pyplot as plt

# load the Excel file into a pandas dataframe
df = pd.read_excel('Assignment.xlsx')
```

```python
# cleaning the dates lables in both trees to show only dates
df.iloc[0:1,1:] = df.iloc[0:1, 1:].apply(lambda s: s.str[8:-1])
df.iloc[22:23,1:] = df.iloc[22:23, 1:].apply(lambda s: s.str[8:-1])
```

## Table number 1 - data for tree #3

```python
#extracting data for tree number 3 and creating new dataframe
dftree3 = df.iloc[1:21]
dftree3.columns =df.iloc[0]
dftree3 = dftree3.set_index('Fruit #')
dftree3.loc['mean of tree #3'] = dftree3.mean()

dftree3
```

## Table number 2 - data for tree #5

```python
#extracting data for tree number 5 and creating new dataframe

dftree5 = df.iloc[23:43]
dftree5.columns =df.iloc[22]
dftree5 = dftree5.set_index('Fruit #')
dftree5.loc['mean of tree #5'] = dftree5.mean()

dftree5
```

## Graph number 1 - a line plot for each fruit on tree #3

```python
for index, row in dftree3.iterrows():
    plt.plot(row, label=index)
plt.xticks(rotation = 60) # Rotates X-Axis Ticks by 60-degrees

plt.xlabel("Date")
plt.ylabel("Size (in mm)")
plt.title("Tree #3\n\n The change in fruit size (in mm) during a period of time")
# plt.legend()
plt.show()
```

### 2.1 Suggest different ways to predict the total number of fruits on the tree

Since the images Fruitspec takes tell us only about the visible fruits, we need an estimation that will complete the blind spots where there are fruits we cant see in the image to predict the total number of fruits on the tree.

The best way to predict the total number of fruits on the tree will be to use image processing as well to determine the total size of the tree, and with that information we can calculate the total volume of the tree. To that we add the estimated distance between each fruit and the way to know that is by first using the fruit recognition we already have and adding to that agronomic knowledge to know better about the distances expected in different parts of the tree. For example branches in the bottom of the tree that are in the shade will probably be less productive than in the top branches exposed to more sunlight. Adding all that knowledge into a three dimensional picture will give us the full estimate of all the locations of the fruits on the tree and the total number of fruits.

Another option is to use machine learning, which is based on a known dataset of images to compare to our own images. Using a known dataset of images with the true count of fruits and then comparing it to our own images can give us a close estimate of the amount of fruits on the tree. The more dataset we have to train our algorithm and taking in account all the variations between the dataset and our images (fruit density, tree size, weather, soil and farmer decisions in the orchard) can give us a good and accurate estimation of the number of fruits on the trees in our orchard.
